<h1><center>RANSAC Algorithm</h1>

<p>  <b>Random Sample Consensus </b>divides data into two subsets : outlier and inlier.For each iteration we randomly select a subset of data as hypothetical inliers <br>
    
The size of the subset should be big enought to fit the model .Example size should be at least n+1 for model having n features.
Ransac algorithm is based on two assumptions: there are enough inlier points to agree on a good model and the outliers will not vote consistently for any single model.This is vital because the outliers will then consistently create their own model and iteration will fall into local optimum.<br>


How many samples should we choose?<br>

N=number of samples<br>
e=prob that a point is an outlier<br>
s=no of points in a sample<br>
p=desired prob that we get a good sample<br>



\begin{align}
\dot{N} & = \log(1-p)/\log(1-(1-e)^s )\\
\end{align}

</p>

    

<p><b>Algorithm:</b><br>
1.Sample a subset of data uniformly at random (the minimum number of points needed to estimate the model<br>
2.Estimate parameters for the model of choice using the sampled subset.<br>
3.Calculate the error for all remaining samples using an error function.<br>
4.Count number of inliers i.e., all samples below a given threshold error.<br>
5.If the number of inliers are above a given threshold, recompute the model using all inliers and hypothesis points<br>
6.Repeat the above to find the best model.<br>
                                               
<b>cons:data loss </b></p>
